```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x_t = (0.25 * (t - 2) ** 2 - 1)*2.7
#y_t = (-0.25 * (t) ** 2)*1.5

time_l = []
x_l = []
y_l = []
for t in np.arange(0.0, 3.3, 0.1):
    time_l.append(t)
    x_l.append((1.7/3.3**2) * (t - 3.3) ** 2 - 1.7)
    y_l.append((-1.45/3.3**2) * t ** 2)
    #t = 3.3 y_l = -1.45
    #t = 0 y_l = 0

plt.plot(time_l,x_l,"b",label = "x")
plt.plot(time_l,y_l,"r",label = "y")
plt.title("x and y")
plt.xlabel("time")
plt.ylabel("m")
plt.legend()
plt.show()
```
