```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#x_t = (0.25 * (t - 2) ** 2 - 1)*2.7
#y_t = (-0.25 * (t) ** 2)*1.5

time_l = []
x_l = []
y_l = []
# add
points_x = [0,-1.7,-2.5,-3.0]
points_y = [0,-1.45,-2.0,-2.5]
time_list = [0,3.3,4,4.7]
step = 0

'''
random_point = (a,b)
top_point = (c,d)
y  = ((b-d)/((a-c)**2)) * ((x-c)**2) + d
'''

def solve(x, a, b, c, d):#d:x_end/y_start,b:y_end/y_start,a:x_time_start/y_time_end,c:x_time_end/y_time_start
    y  = ((b-d)/((a-c)**2)) * ((x-c)**2) + d
    return y

for t in np.arange(0.0, 3.3, 0.1):
    time_l.append(t)
    #x_l.append((1.7/3.3**2) * (t - 3.3) ** 2 - 1.7)
    #y_l.append((-1.45/3.3**2) * t ** 2)
    x_l.append(solve(t, time_list[step], points_x[step], time_list[step+1], points_x[step+1]))
    y_l.append(solve(t, time_list[step+1], points_y[step+1], time_list[step], points_y[step]))
    #t = 3.3 y_l = -1.45
    #t = 0 y_l = 0
step += 1
for t in np.arange(3.3, 4, 0.1):
    time_l.append(t)
    #x_l.append(-0.8/0.7**2 * (t - 3.3) ** 2 - 1.7)
    #t = 4, x_l = -2.5 0.49 0.8
    #y_l.append((2-1.45)/0.7**2 * (t-4) ** 2 - 2)
    x_l.append(solve(t, time_list[step], points_x[step], time_list[step+1], points_x[step+1]))
    y_l.append(solve(t, time_list[step+1], points_y[step+1], time_list[step], points_y[step]))
    #t = 3.3 y_l = -1.45 x_l = -1.7
    #t = 0 y_l = 0

for t in np.arange(4, 4.7, 0.1):
    time_l.append(t)
    x_l.append(solve(t, time_list[step], points_x[step], time_list[step+1], points_x[step+1]))
    y_l.append(solve(t, time_list[step+1], points_y[step+1], time_list[step], points_y[step]))
    #x_l.append(solve(t, 4, -2.5, 4.7, -3))
    #y_l.append(solve(t, 4.7, -2.5, 4, -2))

plt.plot(time_l,x_l,"b",label = "x")
plt.plot(time_l,y_l,"r",label = "y")
plt.title("x and y")
plt.xlabel("time")
plt.ylabel("m")
plt.legend()
plt.grid()
plt.show()
```
